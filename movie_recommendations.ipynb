# Import required libraries

```python
import pandas as pd
```

# Read the dataset

```python
movies=pd.read_csv('reviews.csv')
movies.head()
```

# Attributes of the dataset

```python
movies.columns
```

# Selecting reuquired attributes

```python
movies=movies[['id', 'title', 'overview', 'genre']]
movies
```

```python
movies['tags']=movies['overview']+movies['genre']
movies_as_required=movies.drop(columns=['overview', 'genre'])
movies_as_required
```

# Movie Recommendation

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
vector=cv.fit_transform(movies_as_required['tags'].values.astype('U')).toarray()
similar_movies=cosine_similarity(vector)
```

```python
def recommend_movies(movies):
    index = movies_as_required[movies_as_required['title']==movies].index[0]
    distance = sorted(list(enumerate(similar_movies[index])), reverse=True, key=lambda vector:vector[1])
    for i in distance[1:6]:
        print(movies_as_required.iloc[i[0]].title)
```

```python
recommend_movies("Hulk")
```
